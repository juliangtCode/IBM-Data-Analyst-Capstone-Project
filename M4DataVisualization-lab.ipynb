## Demo: How to run an sql query


```python
# print how many rows are there in the table named 'master'
QUERY = """
SELECT COUNT(*)
FROM master
"""

# the read_sql_query runs the sql query and returns the data as a dataframe
df = pd.read_sql_query(QUERY,conn)
df.head()
```

## Demo: How to list all tables


```python
# print all the tables names in the database
QUERY = """
SELECT name as Table_Name FROM
sqlite_master WHERE
type = 'table'
"""
# the read_sql_query runs the sql query and returns the data as a dataframe
pd.read_sql_query(QUERY,conn)
```

## Demo: How to run a group by query


```python
QUERY = """
SELECT Age,COUNT(*) as count
FROM master
group by age
order by age
"""
pd.read_sql_query(QUERY,conn)
```

## Demo: How to describe a table


```python
table_name = 'master'  # the table you wish to describe

QUERY = """
SELECT sql FROM sqlite_master
WHERE name= '{}'
""".format(table_name)

df = pd.read_sql_query(QUERY,conn)
print(df.iat[0,0])
```

# Hands-on Lab


## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
# your code goes here
%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [7,7]

QUERY = """
SELECT Respondent, ConvertedComp
FROM master
"""
df_ConvertedComp = pd.read_sql_query(QUERY,conn)

# plot histogram
df_ConvertedComp['ConvertedComp'].plot(kind='hist')
plt.title('Converted Compensation')
plt.ylabel('Number of Respondents')
plt.xlabel('Annual, USD');
```

### Box Plots


Plot a box plot of `Age.`


```python
# your code goes here

QUERY = """
SELECT Respondent, Age
FROM master
"""
df_Age = pd.read_sql_query(QUERY,conn)

# plot histogram
df_Age['Age'].plot(kind='box')
plt.title('Box Plot');
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
# your code goes here

QUERY = """
SELECT Age, WorkWeekHrs
FROM master
"""
df = pd.read_sql_query(QUERY,conn)

# plot histogram
df.plot(kind='scatter',x='Age', y = 'WorkWeekHrs')
plt.ylabel('Work Week Hours')
plt.xlabel('Age');
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
# your code goes here

QUERY = """
SELECT CodeRevHrs, WorkWeekHrs, Age
FROM master
"""
df_bubble = pd.read_sql_query(QUERY,conn)

norm_Age = (df_bubble['Age'] - df_bubble['Age'].min()) / (df_bubble['Age'].max() - df_bubble['Age'].min())

df_bubble.plot(kind='scatter', x='WorkWeekHrs', y='CodeRevHrs', alpha=0.3, s=norm_Age*500);
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
# your code goes here

QUERY = """
SELECT * 
FROM DatabaseDesireNextYear
"""
df_db = pd.read_sql_query(QUERY,conn)

df_groupby = df_db.groupby('DatabaseDesireNextYear', axis=0).count()

df_groupby.sort_values(["Respondent"], ascending=False, inplace=True)

df_pie = df_groupby.head()

df_pie['Respondent'].plot(kind='pie', figsize=(5, 6), autopct='%1.1f%%', startangle=90)

plt.title('Top 5 Databases that Respondents Wish to Learn Next Year')
plt.axis('equal');
```

```python
QUERY = """
SELECT * 
FROM LanguageDesireNextYear
"""
df_db = pd.read_sql_query(QUERY,conn)

df_groupby = df_db.groupby('LanguageDesireNextYear', axis=0).count()

df_groupby.sort_values(["Respondent"], ascending=False, inplace=True)

df_pie = df_groupby.head()

df_pie['Respondent'].plot(kind='pie', figsize=(5, 6), autopct='%1.1f%%', startangle=90)

plt.title('Top 5 Languages that Respondents Wish to Learn Next Year')
plt.axis('equal');
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
# your code goes here

QUERY = """
SELECT CodeRevHrs, WorkWeekHrs, Age
FROM master
"""
df_stacked = pd.read_sql_query(QUERY,conn)

df_stacked = df_stacked.groupby('Age', axis=0).median()

df_stacked[30:35].plot(kind='bar', stacked=True)

plt.title('Stacked Chart of Median WorkWeekHrs and CodeRevHrs For The Age Group 30 to 35')
plt.xlabel('Age') # add to x-label to the plot
plt.ylabel('Hours') # add y-label to the plot
plt.title('Median Hours by Age');
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
# your code goes here

QUERY = """
SELECT ConvertedComp, Age
FROM master
"""
df_comp = pd.read_sql_query(QUERY,conn) 

df_comp = df_comp.groupby('Age', axis=0).median()

df_comp[45:60].plot(kind='line')

plt.xlabel('Age')
plt.ylabel('Compensation, USD')
plt.title('Median Compensation by Age');
```

Create a horizontal bar chart using column `MainBranch.`
